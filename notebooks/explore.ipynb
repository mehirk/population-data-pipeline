```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path(os.getcwd()).resolve().parent
file_path = BASE_DIR / "data" / "raw" / "data.csv"


df = pd.read_csv(file_path, skiprows=4)

df.head()
```

```python
df.info()

print(df.columns.tolist())

print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
df_clean = df[["Country Name", "Country Code", "2022"]].copy()
df_clean = df_clean.rename(columns={
    "Country Name": "country",
    "Country Code": "country_code",
    "2022": "population"
})

df_clean.head()
```

```python
df_clean = df_clean.dropna(subset=["population"])
```

```python
df_clean["population"] = df_clean["population"].astype(int)
```

```python
print("Cleaned data shape:", df_clean.shape)
```

```python
BASE_DIR = Path(os.getcwd()).resolve().parent
file_path = BASE_DIR / "data" / "raw" / "data.csv"
output_path = BASE_DIR / "data" / "processed" / "population_clean.csv"
df_clean.to_csv(output_path, index=False)
print(f"Cleaned dataset saved at: {output_path}")
```
